```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Desktop/BRAZIL_CITIES.csv', sep = ";")
```

```python
# Replacing null values with not null values of same data type (either 'object' or 'int' values)
for x in range(len(df.dtypes)):
    if df.dtypes[x] == object:
        df.iloc[:,x].fillna("none", inplace = True)
    else:
        df.iloc[:,x].fillna(0.0, inplace = True)
```
